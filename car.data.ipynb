```python
import csv
with open('car.data','r')as f:
    car_data = csv.reader(f,delimiter = '\n')
    print(car_data)
    query = """insert into cardataset.car(buying,maint,doors,persons,lug_boot,safety,classes) values(%s,%s,%s,%s,%s,%s,%s)"""
    for i in car_data:
        split = str(i[0]).split(",")
        cursor.execute(query, split)
mydb.commit()        
```
